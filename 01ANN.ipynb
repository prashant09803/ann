```python
import numpy as np
import matplotlib.pyplot as plt

x= np.linspace(-5, 5, 100)

def step_func(x):
    step_f=0
    if x>=0:
        step_f = 1
    else:
        step_f = 0
    return step_f
        
def sigmoid_func(x):
    func = 1/(1+np.exp(-x))
    return func

def relu_func(x):
    func = np.maximum(0, x)
    return func
    
def tanh_func(x):
    func = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
    return func
    
def leakyrelu_func(x, a=0.1):
    func = np.where(x>0, x, a*x)
    return func

plt.subplot(2, 3, 1)
plt.plot(x, sigmoid_func(x))
plt.title("Sigmoid")
plt.grid(True)
```

```python
plt.subplot(2, 3, 1)
plt.plot(x, [step_func(y) for y in x])
plt.title("Step_function")
plt.grid(True)
```

```python
plt.subplot(2, 3, 1)
plt.plot(x, relu_func(x))
plt.title("Relu")
plt.grid(True)
```

```python
plt.subplot(2, 3, 1)
plt.plot(x, tanh_func(x))
plt.title("Tan h")
plt.grid(True)
```

```python
plt.subplot(2, 3, 1)
plt.plot(x, leakyrelu_func(x))
plt.title("Leaky Relu")
plt.grid(True)
```
